```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("df_subscribed.csv", index_col=0)
```

```python
X = df.drop("y", axis=1)
```

```python
# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Cluster data using DBSCAN
clustering = DBSCAN(eps=3, min_samples=10).fit(X_scaled)
df["cluster"] = clustering.labels_
```

```python
from sklearn.manifold import TSNE
X_tsne = TSNE(n_components=2).fit_transform(X_scaled)
```

```python
df["x_component"]=X_tsne[:,0]
df["y_component"]=X_tsne[:,1]
```

```python
import plotly.express as px
    
fig = px.scatter(df[df["cluster"]!= -1], x="x_component", y="y_component", color = "cluster", size_max=60)
fig.update_layout(
     height=800)
fig.show()
```

```python
# Calculate statistic per cluster - mean value, Z-score of mean value
cluster_mean = df.groupby("cluster").mean().unstack().reset_index()
cluster_analysis = df.groupby("cluster", group_keys=True).mean().apply(lambda x: (x - x.mean())/ x.mean())
cluster_z_score = cluster_analysis.unstack().reset_index()
cluster_summary = cluster_mean.merge(cluster_z_score, on=["cluster", "level_0"]).rename(columns= {"0_x": "mean", "0_y": "z_score"})
```

```python
cluster_summary[cluster_summary["cluster"]==10].sort_values(by="z_score", ascending=False, key=abs)[:5]
```
